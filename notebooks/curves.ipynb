```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
FREQ = 60
DUR = 40
```

```python
curve = np.zeros(FREQ*DUR + 20)
for i in range(FREQ * DUR):
    x = i / FREQ
    curve[i] = (1/6000)*x**3 - (1/55)*(x-15)**2 - (1/20)*x + 9

for i in range(FREQ*DUR, FREQ*DUR + 20):
    curve[i] = curve[FREQ*DUR - 1]
```

```python
plt.plot(np.linspace(0,DUR+19/FREQ, num=FREQ*DUR+20),curve)
plt.show()
```

```python
curve = np.zeros(FREQ*DUR + 20)
for i in range(FREQ * DUR):
    x = i / FREQ
    curve[i] = 92-(x/20)

for i in range(FREQ*DUR, FREQ*DUR + 20):
    curve[i] = curve[FREQ*DUR - 1]
```

```python
plt.plot(np.linspace(0,DUR+19/FREQ, num=FREQ*DUR+20),curve)
plt.show()
```
